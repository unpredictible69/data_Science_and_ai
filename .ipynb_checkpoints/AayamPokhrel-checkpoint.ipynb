```python
#data types
STR="String"
print("STR: ",STR, type(STR))
integer=10
print("integer: ",integer,type(integer))
floating=10.5
print("floating: ",floating,type(floating))
complex_data=2j
print("complex_data: ",complex_data, type(complex_data))
listed=["Aayam","Pokhrel"]
print("listed: ",listed,type(listed))
tuples=("Aayam","Pokhrel")
print("tuples: ",tuples,type(tuples))
range_value=range(7)
print("range_value: ",range_value,type(range_value))
dictionary={"name":"Aayam","surname":"Pokhrel"}
print("dictionary: ",dictionary,type(dictionary))
set_data={"Aayam","pokhrel"}
print("set_data: ",set_data,type(set_data))
frozen_set=frozenset({"Aayam","Pokhrel"})
print("frozen_set: ",frozen_set,type(frozen_set))
is_it=True
print("is_it: ",is_it,type(is_it))
or_is_it=False
print("is_it: ",or_is_it,type(or_is_it))
bytes_data=b"Aayam?"
print("bytes_data: ",bytes_data,type(bytes_data))
byte_array=bytearray(5)
print("bytearray: ",byte_array,type(byte_array))
memory_view=memoryview(bytes(5))
print("memory_view: ",memory_view,type(memory_view))
none_type=None
print("none_type: ",none_type,type(none_type))
```

```python
# Escape sequences or escape characters
print("Hello \" ' \\world!")
print("Now using escape sequences\n \\n is a line break escape sequence")
print("\n\\\\ is a backslash sequence to write backslash(\\) on print statement")
print("\nTo write \" double quotes, you write \\\" as a double quote sequence")
print("\nTo write ' single quote, you write \\' as a single quote sequence")
print("This is a tab sequence '\\t'\t:")
```
